# Zomato Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
```

```python
dataframe
```

```python
def rate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(rate)
print(dataframe.head())
```

```python
dataframe.info()
```

# 1) Type Of Restaurant do the majority of customers order from?

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of resturant")
```

# Conclusion: Majority of customers order from dinning

```python
dataframe.head()
```

# 2) How many votes has each type of resturant received from customers

```python
groupdata=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':groupdata})
plt.plot(result, c="red", marker="o")
plt.xlabel("Type of restaurant")
plt.ylabel("votes")
```

# Conclusion: Dinning resturant has received maximum votes

```python
dataframe.head()
```

# 3) What are the ratings the majority of resturants have received

```python
plt.hist(dataframe['rate'],bins=15)
plt.title("ratings distribution")
plt.show()
```

#  Conclusion: The majority resturants received ratings from 3.5 to 4

# 4) Zomato has observerd that most couples order most of their food online. What is their average spending on each other

```python
data=dataframe['approx_cost(for two people)']
sns.countplot(x=data)
```

# Conclusion: The majority of couples prefer resturants approx cost of 300 rs.

# 5) Which mode has received the maximum rating?

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

# Conclusion: offline order received low rating in comparision to online order

# 6) Which type of restaurant received more offline orders , so  that Zomato can offer customers with some good offers? 

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title("heatmap")
plt.xlabel("online order")
plt.ylabel("listed in (type)")
plt.show()
```

# Conclusion Dinnning reasturents primarily accept offline orders,whereas cafes primarily receive online orders
